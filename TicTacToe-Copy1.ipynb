```python
#Tic Tac Toe game in python
### hold the board, check the move possibilities
class startGame:
    def __init__(self):
        self.board = [' ' for x in range(10)]
        self.P1 = CompPlayer()
        self.P2 = CompPlayer()
        
    def printBoard(self, board):
        print(' ' + self.board[1] + ' | ' + self.board[2] + ' | ' + self.board[3])
        print('-----------')
        print(' ' + self.board[4] + ' | ' + self.board[5] + ' | ' + self.board[6])
        print('-----------')
        print(' ' + self.board[7] + ' | ' + self.board[8] + ' | ' + self.board[9])
        
    def insertLetter(self, letter, pos):
        self.board[pos] = letter

    def spaceIsFree(self, pos):
        return self.board[pos] == ' '

    def isWinner(self, bo, le):
        return (bo[7] == le and bo[8] == le and bo[9] == le) or (bo[4] == le and bo[5] == le and bo[6] == le) or(bo[1] == le and bo[2] == le and bo[3] == le) or(bo[1] == le and bo[4] == le and bo[7] == le) or(bo[2] == le and bo[5] == le and bo[8] == le) or(bo[3] == le and bo[6] == le and bo[9] == le) or(bo[1] == le and bo[5] == le and bo[9] == le) or(bo[3] == le and bo[5] == le and bo[7] == le)

    def isBoardFull(self, board):
        if self.board.count(' ') > 1:
            return False
        else:
            return True

    def main(self):
        print('Welcome to Tic Tac Toe!')
        self.printBoard(self.board)

        while not(self.isBoardFull(self.board)):
            if not(self.isWinner(self.board, 'O')):
                try:
                    self.P1move = int(self.P1.makeMove())
                except ValueError or EOFError as e:
                    print(e)
                    continue
                else:
                    self.insertLetter('X', self.P1move)
                    self.P2.getMove("X",self.P1move)
                    self.P1.insertLetter('X', self.P1move)
                    self.printBoard(self.board)
                
                #self.P1move = self.P1.makeMove()
                #self.insertLetter('X', self.P1move)
                #self.P2.getMove("X",self.P1move)
                #self.printBoard(self.board)
                
            else:
                print('Sorry, O\'s won this time!')
                break

            if not(self.isWinner(self.board, 'X')):
                self.P2move = self.P2.makeMove()
                if self.P2move == 0:
                    print('Tie Game!')
                else:
                    self.P1.getMove("O",self.P2move)
                    self.insertLetter('O', self.P2move)
                    print('Computer placed an \'O\' in position', self.P2move , ':')
                    self.P2.insertLetter('O', self.P2move)
                    self.printBoard(self.board)
            else:
                print('X\'s won this time! Good Job!')
                break

        if self.isBoardFull(self.board):
            print('Tie Game!')
```

```python
## instance of player
class Player:
    def __init__(self):
        self.board = [' ' for x in range(10)]
    def insertLetter(self, letter, pos):
        self.board[pos] = letter
    def spaceIsFree(self, pos):
        return self.board[pos] == ' '
    def makeMove(self):
        pass
    def getMove(self,letter,getmove):
        self.insertLetter(letter, getmove)
    def selectRandom(self, li):
        import random
        ln = len(li)
        r = random.randrange(0,ln)
        return li[r]
```

```python
class CompPlayer(Player):
    def __init__(self):
        super().__init__() 
    def makeMove(self):
        print("CompPlayerBoard: ",self.board)
        possibleMoves = [x for x, letter in enumerate(self.board) if letter == ' ' and x != 0]
        self.move = 0
        for let in ['O', 'X']:
            for i in possibleMoves:
                boardCopy = self.board[:]
                boardCopy[i] = let
                if startGame().isWinner(boardCopy, let):
                    self.move = i
                    return self.move

        cornersOpen = []
        for i in possibleMoves:
            if i in [1,3,7,9]:
                cornersOpen.append(i)

        if len(cornersOpen) > 0:
            self.move = self.selectRandom(cornersOpen)
            return self.move

        if 5 in possibleMoves:
            self.move = 5
            return self.move
        
        edgesOpen = []
        for i in possibleMoves:
            if i in [2,4,6,8]:
                edgesOpen.append(i)

        if len(edgesOpen) > 0:
            self.move = self.selectRandom(edgesOpen)
            
        return self.move
```
